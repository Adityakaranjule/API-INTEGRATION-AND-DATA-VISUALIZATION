```python
import requests
import matplotlib.pyplot as plt
```

```python
url=('http://api.weatherapi.com/v1/current.json?key=ba319d0bc6484b2e9f0144601250912&q=Mumbai&aqi=no')
```

```python
data=requests.get(url)
```

```python
Data=data.json()
```

```python
print(Data)
```

```python
temp=Data['current']['temp_c']
feellike=Data['current']['feelslike_c']
humidity=Data['current']['humidity']
dew=Data['current']['dewpoint_c']
```

```python
plt.bar(["temp","feellike","humidity","dew"],[temp,feellike,humidity,dew])
```
